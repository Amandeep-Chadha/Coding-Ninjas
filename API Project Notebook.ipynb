## Qset 1:
Querying on Movies

```python
#Q1: Find id of movie "Andhadhun" using TMDb API.
import requests

url = "https://api.themoviedb.org/3/search/movie?query=Andhadhun&include_adult=false&language=en-US&page=1"

headers = {
    "accept": "application/json",
    "Authorization": "token"
}

response = requests.get(url, headers=headers)

print(response.json()['results'][0]['id'])

#Q2: Fetch the company id of company 'Marvel Studios' suing TMDb.
url = "https://api.themoviedb.org/3/search/company?query=marvel&page=1"

headers = {
    "accept": "application/json",
    "Authorization": "token"
}

response = requests.get(url, headers=headers)

for i in (response.json()['results']):
    if i['name'] == 'Marvel Studios':
        print(i['id'])

#Q3 Find the vote count and vote average of the movie "3 Idiiots" using the TMDb API.
import requests

url = "https://api.themoviedb.org/3/movie/20453?language=en-US"

headers = {
    "accept": "application/json",
    "Authorization": "token"
}

response = requests.get(url, headers=headers)

print(response.json()['vote_count'], response.json()['vote_average'])

#Q4 Fetch the top rated english movies in the US region using the TMDb API. From the result, print the first 10 movies which have original language as english. Also print their genres.
import requests
import pandas as pd

url = "https://api.themoviedb.org/3/movie/top_rated?language=en-US&page=1&region=US"

headers = {
    "accept": "application/json",
    "Authorization": "token"
}

response = requests.get(url, headers=headers)

results = []
item = 0
while len(results) < 10:
    
    if (response.json()['results'][item]['original_language']) == 'en':
        results.append(response.json()['results'][item])
    item += 1

url = "https://api.themoviedb.org/3/genre/movie/list?language=en"

headers = {
    "accept": "application/json",
    "Authorization": "token"
}

response = requests.get(url, headers=headers)

df1 = pd.DataFrame(response.json()['genres'])
df1.index = df1.id
df1.drop(columns=['id'], inplace = True)
for i in results:
    print(i['original_title'], ' -', end = ' ')
    for j in i['genre_ids']:
        print(df1.loc[j][0], end = ', ')
    print('')
```

```python
#Q1 Fetch the overvuew of the TV Show. "FRIENDS" using TMDb API.

import requests

url = "https://api.themoviedb.org/3/search/tv?query=FRIENDS&include_adult=false&language=en-US&page=1"
headers = {
    "accept": "application/json",
    "Authorization": "token"
}

response = requests.get(url, headers=headers)

print(response.json()['results'][0]['overview'])
print(response.json()['results'][1]['overview'])


#Q2 Fetch the name and air date of S06E05 of the TV Show, 'The Big Bang Theory' form TMDb API.
import requests

url = "https://api.themoviedb.org/3/tv/1418/season/6?language=en-US"

headers = {
    "accept": "application/json",
    "Authorization": "token"
}

response = requests.get(url, headers=headers)

for i in (response.json()['episodes']):
    if i['episode_number'] == 5:
        print(i['name']," - ", i['air_date'])
        
#Q3 Fetch the trending TV Shows for the week from the TMDb API and print the taglines of the top 5 shows. If there is no tagline, print 'Empty' instead

import requests

url = "https://api.themoviedb.org/3/trending/tv/week?language=en-US"

headers = {
    "accept": "application/json",
    "Authorization": "token"
}
response = requests.get(url, headers=headers)
results = []
for i in range(5):
    results.append(response.json()['results'][i]['id'])

for i in results:
    url = "https://api.themoviedb.org/3/tv/{0}?language=en-US".format(i)
    response = requests.get(url, headers=headers)
    if not response.json()['tagline']:
        print('Empty')
    else:
        print(response.json()['tagline'])
```
