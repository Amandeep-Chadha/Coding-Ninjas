```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
import warnings
warnings.filterwarnings("ignore")
```

```python
training = np.loadtxt('./training_data.csv', delimiter = ',')
```

```python
def prepare(training):
    df = pd.DataFrame(training)
    for i in range(13):
        for j in range(i, 13):
            for k in range(j, 13):
                df[f'{i}_{j}_{k}'] = df[i]*df[j]*df[k]
    df['C'] = 1
    col_list = list(df.columns)
    col_list[13], col_list[-2] = col_list[-2], col_list[13]
    col_list[-2], col_list[-1] = col_list[-1], col_list[-2]
    df = df[col_list]
    training = np.array(df) 
    x = training[:,:-1]
    y = training[:,-1]
    return x, y
def step_grad(m, x, y, step_length):
    m_slope = np.zeros(len(x[0]))
    M = len(x[0])
    for i in range(len(x)):
        xi = x[i]
        yi = y[i]
        m_slope += (-2/M)*(yi-(m*xi).sum())*(xi)
    return (m - (step_length*m_slope))
def cost(m, x, y):
    cost=0
    for i in range(len(x)):
        cost +=(1/len(x))*((y[i]-(m*x[i]).sum())**2)
    return(cost)
def gd(x, y, step_length, num_iter):
    m = np.zeros(len(x[0]))
    for i in range(num_iter):
        m = step_grad(m, x, y, step_length)
        print(f'Cost for iteration {i}: {cost(m, x, y)}')
    return m
def run(x, y):
    num_iter = 300
    step_length = 0.000268
    m = gd(x, y, step_length, num_iter)
    print(m[:14])
    return m
```

```python
training = np.loadtxt('./training_data.csv', delimiter = ',')
x, y = prepare(training)


m = run(x, y)
pred = []

for i in range(len(x)):
    sigma = np.append(x[i,:-1], np.ones(1))
    pred.append(sum(sigma*m))
    
print(pred)
```

```python
training = np.loadtxt('./training_data.csv', delimiter = ',')
x, y = prepare(training)
df = pd.DataFrame(x)
standard_scaler = preprocessing.StandardScaler()
standard_scaler.fit(df.iloc[:,:-1])
df = standard_scaler.transform(df.iloc[:,:-1])
for i in range(len(df)):
    x[i] = np.append(df[i], np.ones(1))
m = run(x, y)
pred = []

for i in range(len(x)):
    sigma = np.append(df[i,:], np.ones(1))
    pred.append(sum(sigma*m))
print(pred)
```
