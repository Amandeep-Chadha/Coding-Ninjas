## Question 1

Finding top 3 cities with highest number of investments

```python
#included to ignore warning raised while changing element values by boolean indexing
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#reading dataset
df = pd.read_csv('startup_funding.csv')
df.dropna(subset=['City  Location'], inplace = True)
cities = np.array([])


def cleaning(name):
    if name == 'Delhi' or name == 'Noida' or name == 'Gurgaon' or 'Delhi' in name or name == 'New Delhi':
        return 'Delhi-NCR'
    elif name == 'bangalore':
        return 'Bangalore'
    else:
        return name


for i in df['City  Location']:
    lis = i.split('/')
    for j in lis:
        j = j.strip()
        cities = np.append(cities, j)
city = pd.Series(cities)
city = city.apply(cleaning)
counts = city.value_counts().head(3)
print("The top three cities where investors invest are:")
display(counts)
print("Refer to the plot below:")
plt.bar(counts.index, counts)
plt.show()
```

### Solution: Delhi-NCR is the best place to start as it has the highest amount of startups funded as seen in the city series and the plot

## Question 2

Finding top 5 investors(frequency of investments)

```python
import pandas as pd
import numpy as np


df = pd.read_csv('startup_funding.csv', encoding = 'utf-8')
df.dropna(subset = ['Investors Name'], inplace = True)
startups = pd.array([])

for i in df['Investors Name']:
    lis = i.split(',')
    for i in lis:
        if i == "":
            continue
        else:
            i = i.strip()
            startups = np.append(startups, i)

startarray = pd.Series(startups)
df1 = (startarray.value_counts())
plt.bar(df1.index[:5], df1[:5])
plt.xticks(rotation=30)
plt.show()

for i in df1.index[0:5]:
    print(i)
```

### Solution: 
--Sequoia Capital 
--Accel Partners
--Kalaari Capital
--Blume Ventures
--SAIF Partners

## Question 3

Finding top 5 investors(investment frequency in different stratups).

```python
import pandas as pd
import numpy as np


df = pd.read_csv('startup_funding.csv', encoding = 'utf-8')
df.dropna(subset = ['Investors Name'], inplace = True)
df['Startup Name'].replace('Ola Cabs', 'Ola', inplace = True)
df['Startup Name'].replace('Flipkart.com', 'Flipkart', inplace = True)
df['Startup Name'].replace('Oyorooms', 'Oyo', inplace = True)
df['Startup Name'].replace('OYO Rooms', 'Oyo', inplace = True)
df['Startup Name'].replace('Oyo Rooms', 'Oyo', inplace = True)
df['Startup Name'].replace('OyoRooms', 'Oyo', inplace = True)
df['Startup Name'].replace('OYO', 'Oyo', inplace = True)
df['Startup Name'].replace('OYO', 'Oyo', inplace = True)
df['Startup Name'].replace('Olacabs', 'Ola', inplace = True)
df['Startup Name'].replace('Paytm Marketplace','Paytm',inplace=True)
companies = pd.unique(df['Startup Name'])
df1 = pd.DataFrame(columns = ['Startup', 'Investor'])
index = 0
for i in companies:
    for j in df['Investors Name'][df['Startup Name'] == i]:
        lis = j.split('/')
        for k in lis:
            k = k.strip()
            df1.loc[index] = [i, k]
            index+=1
print(df1)
```

```python
df1 = df1.drop_duplicates(keep = 'first')
for i in df1['Investor']:
    lis = i.split(',')
    for i in lis:
        if i == "":
            continue
        else:
            i = i.strip()
            startups = np.append(startups, i)

startarray = pd.Series(startups)
df2 = (startarray.value_counts())
plt.bar(df2.index[:5], df2[:5])
plt.xticks(rotation=30)
plt.show()

for i in df2.index[0:5]:
    print(i)
```

### Solution: 
-- Sequoia Capital
-- Accel Partners
-- Kalaari Capital
-- SAIF Partners
-- Blume Ventures
